```python
#Question continous distribution(a)
#(i)uniform distribution(CDF AND PDF)

from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

a, b = 0, 1
c, d = 2, 2
e, f = 4, 3
size = 1000

fig, ax = plt.subplots(1, 1)
fig, ax1 = plt.subplots(1, 1)
fig, ax2 = plt.subplots(1, 1)

# Genrating uniform distribution
uniform_distribution = uniform(loc=a, scale=b)
uniform_distribution1 = uniform(loc=c, scale=d)
uniform_distribution2 = uniform(loc=e, scale=f)
x = np.linspace(uniform_distribution.ppf(0), uniform_distribution.ppf(1), size)
y = np.linspace(uniform_distribution1.ppf(0), uniform_distribution1.ppf(1), size)
z = np.linspace(uniform_distribution2.ppf(0), uniform_distribution2.ppf(1), size)

# Plotting pdf
pdf = uniform_distribution.pdf(x)
pdf = uniform_distribution1.pdf(y)
pdf = uniform_distribution2.pdf(z)
ax.plot(x, pdf, 'r-', lw=5, alpha=0.6, label='pdf')
ax1.plot(y, pdf, 'r-', lw=5, alpha=0.6, label='pdf')
ax2.plot(z, pdf, 'r-', lw=5, alpha=0.6, label='pdf')

# Plotting cdf
cdf = uniform_distribution.cdf(x)
cdf = uniform_distribution1.cdf(y)
cdf = uniform_distribution2.cdf(z)
ax.plot(x, cdf, 'k-', lw=2, label='cdf')
ax1.plot(y, cdf, 'k-', lw=2, label='cdf')
ax2.plot(z, cdf, 'k-', lw=2, label='cdf')
ax.legend(loc='best', frameon=False)
ax1.legend(loc='best', frameon=False)
ax2.legend(loc='best', frameon=False)







     
        
```

```python

#(ii)Exponential distribution

import scipy.stats as ss

def plot_exponential(x_range, mu=0, sigma=1, cdf=False, **kwargs):
    '''
    Plots the exponential distribution function for a given x range
    If mu and sigma are not provided, standard exponential is plotted
    If cdf=True cumulative distribution is plotted
    Passes any keyword arguments to matplotlib plot function
    '''
    x = x_range
    if cdf:
        y = ss.expon.cdf(x, mu, sigma)
    else:
        y = ss.expon.pdf(x, mu, sigma)
    plt.plot(x, y, **kwargs)
    
x = np.linspace(0, 5, 5000)
plot_exponential(x, -2, 1, color='red', lw=2, ls='-', alpha=0.5, label='pdf')
plot_exponential(x, 2, 1.2, color='red', lw=2, ls='-', alpha=0.5, label='pdf')
plot_exponential(x, 0, .8, color='red', lw=2, ls='-', alpha=0.5, label='pdf')
plot_exponential(x, -2, 1, cdf=True, color='blue', lw=2, ls='-', alpha=0.5, label='cdf')
plot_exponential(x, 2, 1.2, cdf=True, color='blue', lw=2, ls='-', alpha=0.5, label='cdf')
plot_exponential(x, 0, .8, cdf=True, color='blue', lw=2, ls='-', alpha=0.5, label='cdf')
plt.legend();
```

```python
#discrete distribution(a)

#It has two possible outcome posiible in single trail so it's a bernoulli distribution in which
#  P(X=x)=   p; for x=0(failure)
#          1-p; for x=1(success)

#Here we have given p(failure)=0.1,0.5,0.8
#(i)For p=0.1
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
#
# Instance of Bernoulli distribution with parameter (1-p=q)  
#
bd = bernoulli(0.9)
#
# Outcome of experiment can take value as 0, 1
#
X = [0, 1]
#bernoulli probability mass function:
def pmf(x,q):
    f = q**x*(1-q)**(1-x)
    return f
# expected value of x
def mean(q):
    return q
#variance of x
def var(q):
    return q*(1-q)
# Create a bar plot; Note the usage of "pmf" function
# to determine the probability of different values of
# random variable
#
plt.figure(figsize=(7,7))
plt.xlim(-1, 2)
plt.bar(X, bd.pmf(X), color='orange')

plt.title('Bernoulli Distribution (p=0.1)', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
print("mean: ",mean(0.9))
print("variance: ",var(0.9))
print("PMF at x=0 i.e. failure: ",pmf(0,0.9))
print("PMF at x=1 i.e. failure: ",pmf(1,0.9))

```
